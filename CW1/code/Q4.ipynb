```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
data_path = os.path.join(os.getcwd(), '../data', 'classification_eval_1.csv')
df = pd.read_csv(data_path, delimiter = ',')
```

```python
df_binary = df.copy(deep = True)
df_binary[df_binary >= 0.5] = 1
df_binary[df_binary < 0.5] = 0
```

```python
acc_1 = accuracy_score(df_binary['gt'], df_binary['alg_1'])
acc_2 = accuracy_score(df_binary['gt'], df_binary['alg_2'])
acc_3 = accuracy_score(df_binary['gt'], df_binary['alg_3'])
acc_4 = accuracy_score(df_binary['gt'], df_binary['alg_4'])
```

```python
print(acc_1,acc_2,acc_3,acc_4)
```

```python
auc_1 = roc_auc_score(df['gt'], df['alg_1'])
auc_2 = roc_auc_score(df['gt'], df['alg_2'])
auc_3 = roc_auc_score(df['gt'], df['alg_3'])
auc_4 = roc_auc_score(df['gt'], df['alg_4'])
```

```python
print(auc_1,auc_2,auc_3,auc_4)
```

```python
roc_1 = roc_curve(df['gt'], df['alg_1'])
roc_2 = roc_curve(df['gt'], df['alg_2'])
roc_3 = roc_curve(df['gt'], df['alg_3'])
roc_4 = roc_curve(df['gt'], df['alg_4'])
```

```python
plt.title('ROC curves for four models')
plt.grid(linestyle=':')

plt.plot(roc_1[0], roc_1[1], c='orange', label='alg_1')
plt.plot(roc_2[0], roc_2[1], c='lightblue', label='alg_2')
plt.plot(roc_3[0], roc_3[1], c='lightgreen', label='alg_3')
plt.plot(roc_4[0], roc_4[1], c='pink', label='alg_4')
plt.tight_layout()
plt.legend()

plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.show()
```
